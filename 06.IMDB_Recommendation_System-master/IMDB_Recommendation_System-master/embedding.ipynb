```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
sent
```

```python
### One Hot Representation: create a list representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

for example for the sentence 'the glass of milk', in indices 4356, 2252, 3703, 6076, I will get value 1 but the rest of list is set to 0.

```python
## word Embedding Representation

from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

We need to make all the sentences of the same size. So, we are using pad sequences to make them the same size.

if I print padding = 'pre' I will get all the zeros at initial indices, and if I print padding = 'post' I will get them at the last indices.

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_length))
model.build(input_shape=(None, sent_length)) # Explicitly build the model
model.compile('adam', 'mse') # optimizer, loss function
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
